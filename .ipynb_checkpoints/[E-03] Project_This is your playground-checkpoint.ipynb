#### 라이브러리, 데이터 불러오기

```python
import warnings
warnings.filterwarnings("ignore")

import os
from os.path import join

import pandas as pd
import numpy as np

import missingno as msno

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_score
import xgboost as xgb
import lightgbm as lgb

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_dir = 'data/'

train_data_path = join(data_dir, 'train.csv')
test_data_path = join(data_dir, 'test.csv') 

train = pd.read_csv(train_data_path)
test = pd.read_csv(test_data_path)

train.head()
```

```python
msno.matrix(train)
```

데이터를 처리하는 과정에서 추가로 처리를 해야하는 부분이 있는지 확인하기 위해 missingno 라이브러리를 이용해 결측치를 확인해 보았다.

```python
train['date'] = train['date'].apply(lambda i: i[:6]).astype(int)
train.head()
```

```python
y = train['price']
del train['price']

print(train.columns)
```

train의 price 칼럼은 맞춰야하는 target 이므로 이를 y 변수로 지정한 후 train에서 삭제시킨다.

```python
del train['id']

print(train.columns)
```

```python
fig, ax = plt.subplots(9, 2, figsize=(12, 50))

count = 1
columns = train.columns
for row in range(9):
    for col in range(2):
        sns.kdeplot(data=train[columns[count]], ax=ax[row][col])
        ax[row][col].set_title(columns[count], fontsize=15)
        count += 1
        if count == 19:
            break
```

각 칼럼들을 시각화하여 데이터의 분포를 살펴보자. 현재 'bedrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_lot15', 'sqft_living15'의 칼럼들이 왼쪽으로 치우친 것을 볼 수 있다. 이러한 값들은 예측시 결과에 영향을 미칠 수 있기 때문에 이 값들을 log값으로 취해줌으로써 정규분포와 유사한 형태로 만들어주자.

```python
skew_columns = ['bedrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_lot15', 'sqft_living15']

for c in skew_columns:
    train[c] = np.log1p(train[c].values)
```

아래의 그림을 보면 왼쪽으로 치우쳤던 그래프들이 가운데로 모임을 확인할 수 있다.

```python
fig, ax = plt.subplots(4, 2, figsize=(12, 24))

count = 0
for row in range(4):
    for col in range(2):
        if count == 7:
            break
        sns.kdeplot(data=train[skew_columns[count]], ax=ax[row][col])
        ax[row][col].set_title(skew_columns[count], fontsize=15)
        count += 1
```

```python
cor_mat = train.corr()
mask = np.array(cor_mat)
mask[np.tril_indices_from(mask)] = False

flg = plt.gcf()
plt.figure(figsize = (15,15))
sns.heatmap(data = cor_mat, mask = mask, square = True, annot = True, cbar = True)
```

각 칼럼들의 관계를 확인해보기 위해 correlation을 구해보았다. 1에 가까운 값일수록 양의 상관관계를, -1에 가까울 수록 음의 상관관계를 가지게 된다.

```python
test['date'] = test['date'].apply(lambda i: i[:6]).astype(int)

del test['id']

print(test.columns)
```

```python
skew_columns = ['bedrooms', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_lot15', 'sqft_living15']

for c in skew_columns:
    test[c] = np.log1p(test[c].values)
```

```python
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(x.values)
    rmse= np.sqrt(-cross_val_score(model, x.values, y, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
model_xgb = xgb.XGBRegressor(colsample_bytree=0.4, gamma=0.0468, 
                             learning_rate=0.05, max_depth=6, 
                             min_child_weight=1.7817, n_estimators=15000,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.8, silent=1,
                             random_state =25, nthread = -1)
model_lgb = lgb.LGBMRegressor(objective='regression',num_leave=2,
                              learning_rate=0.05, n_estimators=15000,
                              max_bin = 80, bagging_fraction = 0.8,
                              bagging_freq = 5, feature_fraction = 0.2319,
                              feature_fraction_seed=9, bagging_seed=9,
                              min_data_in_leaf =6, min_sum_hessian_in_leaf = 11)
model_gb = GradientBoostingRegressor(n_estimators=15000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =25)
```

```python
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
def rmse_expm1(y, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y), np.expm1(y_pred)))
```

```python
model_xgb.fit(train, y)
xgb_train_pred = model_xgb.predict(train)
xgb_pred = model_xgb.predict(test)
xgb_pred = np.expm1(xgb_pred)
print(rmse_expm1(y, xgb_train_pred))
```

```python
model_lgb.fit(train, y)
lgb_train_pred = model_lgb.predict(train)
lgb_pred = model_lgb.predict(test)
lgb_pred = np.expm1(lgb_pred)
print(rmse_expm1(y, lgb_train_pred))
```

```python
model_gb.fit(train, y)
gb_train_pred = model_gb.predict(train)
gb_pred = model_gb.predict(test)
gb_pred = np.expm1(gb_pred)
print(rmse_expm1(y, gb_train_pred))
```

```python
print('RMSE score on train data:')
print(rmse_expm1(y,xgb_train_pred*0.4 + lgb_train_pred*0.4 + gb_train_pred*0.2 ))
```

```python
ensemble = xgb_pred*0.4 + lgb_pred*0.4 + gb_pred*0.2
```

```python
submission_path = join('data/sample_submission.csv')
submission = pd.read_csv(submission_path)
submission.head()
```

```python
submission['price'] = ensemble
submission.head()
```

```python
submission_csv_path = '{}/submission_{}_RMSE_{}.csv'.format(data_dir, 'lgbm', '0.161525')
submission.to_csv(submission_csv_path, index=False)
print(submission_csv_path)
```
